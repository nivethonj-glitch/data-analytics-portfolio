**TECHNICAL HIGHLIGHTS: DATA ENGINEERING & BEHAVIORAL MODELLING**:

To move beyond surface-level metrics, we utilized Python to engineer a high-fidelity dataset that uncovers the hidden relationship between user activity and platform fatigue.

**Custom Behavioral Archetyping:** We engineered a multi-variable logic to segment the user base into four distinct behavioral clusters: Power Users, Efficient Engagers, Passive Scrollers, and At-Risk users.

**Predictive Churn Risk Scoring:**  We developed a churn propensity model (0.0 to 1.0) that revealed a critical "Quality Flight" trend: Efficient Engagers currently exhibit the highest risk of dormancy at approximately 0.88, a risk factor invisible to traditional engagement tracking.

**EQE (Engagement Quality Efficiency) Metric:** Using Python, we defined and calculated EQE to quantify the value density of user sessions. This metric statistically confirmed an "L-Curve" behavior, proving that engagement quality drops significantly as raw activity volume increases.

**Fatigue Analysis via Ad-Load Proxy:** We calculated a normalized ad_load_proxy to measure monetization pressure. Our findings indicate that Efficient Engagers are currently subjected to nearly 3x the monetization load compared to Passive Scrollers, identifying the primary catalyst for high-value user burnout.

**Forming Correlation Matrix:** We used this to check the relationship between lifestyle factors to see if it can affect churn .

**DEFINING METRICS:**

```python
# 1. Define Meaningful Engagement
df['ME_Score'] = (df['likes_given_per_day'] * 1 +
                  df['comments_written_per_day'] * 3 +
                  (df['dms_sent_per_week'] / 7) * 5 +
                  (df['posts_created_per_week'] / 7) * 10)

# 2. Define Total Attention
df['Total_Attention'] = (df['time_on_feed_per_day'] +
                         df['time_on_reels_per_day'] +
                         df['time_on_explore_per_day'] +
                         df['time_on_messages_per_day'])

# 3. Calculating (EQE)
df['EQE'] = df['ME_Score'] / (df['Total_Attention'] + 0.01)
```

**HANDLING MISSING VALUES & OUTLIERS:**

```python
# Handle missing values
df = df.fillna(df.median(numeric_only=True))

# Remove outliers using IQR for EQE
Q1 = df['EQE'].quantile(0.25)
Q3 = df['EQE'].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df['EQE'] < (Q1 - 1.5 * IQR)) | (df['EQE'] > (Q3 + 1.5 * IQR)))]

# Export for Tableau
df.to_csv("instagram_cleaned_for_tableau.csv", index=False)
```

**DATA INSPECTION:**

```python
import pandas as pd
import gdown
import numpy as np

#LOAD DATA
file_id = "16nG7VaTdxL1xpi3ZuYrUf5lTV7bvTyxP"
download_url = f"https://drive.google.com/uc?id={file_id}"
output_file = "instagram_dataset.csv"


gdown.download(download_url, output_file, quiet=True)
data = pd.read_csv(output_file)

# DATA INSPECTION
print("### DATASET OVERVIEW ###")
print(data.info())

print("\n### LIFESTYLE & USAGE CHECK ###")
# Checking for impossible values
cols_to_check = ['sleep_hours_per_night', 'daily_active_minutes_instagram', 'age']
print(data[cols_to_check].describe())

# Checking for the "Zero Attention" problem that breaks EQE
zero_attention = data[data['daily_active_minutes_instagram'] == 0].shape[0]
print(f"\nUsers with 0 minutes activity: {zero_attention}")

# FEATURE ENGINEERING
# 1. Meaningful Engagement (ME)
data['ME_Score'] = (
    data['likes_given_per_day'] * 1 +
    data['comments_written_per_day'] * 3 +
    (data['dms_sent_per_week'] / 7) * 5 +
    (data['posts_created_per_week'] / 7) * 10
)

# 2. Total Attention (A)
data['Total_A'] = (
    data['time_on_feed_per_day'] +
    data['time_on_reels_per_day'] +
    data['time_on_explore_per_day'] +
    data['time_on_messages_per_day']
)

# 3. Engagement Quality Efficiency (EQE)
data['EQE'] = data['ME_Score'] / (data['Total_A'] + 1)

print("\n### CALCULATION COMPLETE ###")
print(data[['ME_Score', 'Total_A', 'EQE']].head())
```

**USER SEGMENTATION:**

```python

eqe_median = data['EQE'].median()
attention_median = data['Total_A'].median()

def segment_user(row):
    if row['EQE'] >= eqe_median and row['Total_A'] >= attention_median:
        return 'Power User'
    elif row['EQE'] < eqe_median and row['Total_A'] >= attention_median:
        return 'Passive Scroller'
    elif row['EQE'] >= eqe_median and row['Total_A'] < attention_median:
        return 'Efficient Engager'
    else:
        return 'At-Risk'

data['User_Segment'] = data.apply(segment_user, axis=1)


print(data['User_Segment'].value_counts())
```

**CHURN RISK SCORE:**

```python

data['session_risk'] = 1 - (data['sessions_per_day'] / data['sessions_per_day'].max())
data['notif_risk'] = 1 - data['notification_response_rate']


data['Churn_Risk_Score'] = (data['session_risk'] + data['notif_risk']) / 2

print("\nTop 5 At-Risk Users (by score):")
print(data[['user_id', 'User_Segment', 'Churn_Risk_Score']].sort_values(by='Churn_Risk_Score', ascending=False).head())
```

**CORRELATION MATRIX:**

```python

lifestyle_cols = ['sleep_hours_per_night', 'perceived_stress_score', 'exercise_hours_per_week', 'EQE']
correlation_matrix = data[lifestyle_cols].corr()

print("\n### Correlation with Engagement Quality (EQE) ###")
print(correlation_matrix['EQE'].sort_values(ascending=False))
```
